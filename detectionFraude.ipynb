```python
from faker import Faker

import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns

from tqdm import tqdm
import random
```

```python
fake = Faker()

customer_list = []

for i in tqdm(range (500)):
    customer_id = random.choices('0123456789')
    first_name = fake.first_name()
    last_name = fake.last_name()
    domain = fake.domain_name()
    email = f"{last_name.lower()}.{first_name.lower()}@{domain}"
    phone_number = "+237 6" + "".join(random.choices('0123456789', k=8))
    registration_date = fake.date_time_between(start_date = '-4y', end_date = 'now')
    date_of_birth = fake.date_of_birth(minimum_age=17, maximum_age=90)
    address = fake.address()
    city = fake.city()
    region = fake.state()

    customer_dict = {
        "customer_id": customer_id,
        "first_name": first_name,
        "last_name": last_name,
        "email": email,
        "phone_number": phone_number,
        "registration_date": registration_date,
        "date_of_birth": date_of_birth,
        "address": address,
        "city": city,
        "region": region
    }

    customer_list.append(customer_dict)

df = pd.DataFrame(customer_list)

df.head()
```

```python
transaction_list = []

for i in tqdm(range (500)):
    transaction_id = ''.join(random.choices('0123456789', k=1)) 
    transaction_type = random.choice(['sending', 'receiving', 'purchase', 'refund','transfer'])
    amount = round(random.uniform(10, 1000), 2)  # Generate random amount between 10 and 1000
    transaction_date = fake.date_time_between(start_date='-4y', end_date='now')
    location = fake.city()
    device_id = ''.join(random.choices('0123456789', k=8)) 
    status = random.choice(['success', 'failed', 'pending'])
    is_fraud = random.choice([True, False])  # Generate random fraud flag

    transaction_dict = {
        "transaction_id": transaction_id,
        "customer_id": customer_id,
        "transaction_type": transaction_type,
        "amount": amount,
        "transaction_date": transaction_date,
        "location": location,
        "device_id": device_id,
        "status": status,
        "is_fraud": is_fraud
    }

    transaction_list.append(transaction_dict)

df = pd.DataFrame(transaction_list)

df.head()
```

```python
device_list = []

for i in tqdm(range (500)):
    device_id = ''.join(random.choices('0123456789', k=1)) 
    device_type = fake.random_element(elements=('smartphone', 'tablet', 'laptop', 'desktop'))
    os = fake.random_element(elements=('iOS', 'Android', 'Windows', 'macOS'))
    manufacturer = fake.company()
    model = fake.word()
    last_used = fake.date_time_between(start_date='-1y', end_date='now')

    device_dict = {
        "device_id" : device_id,
        "device_type" : device_type,
        "os" : os,
        "manufacturer" : manufacturer,
        "model" : model,
        "last_used" : last_used 
    }
    
    device_list.append(device_dict)

df = pd.DataFrame(device_list)

df.head()
```

```python
transHistory_list = []

for i in tqdm(range (500)):
    history_id = fake.unique.random_int()
    previous_balance = round(random.uniform(100, 1000), 2)
    new_balance = previous_balance + round(random.uniform(-500, 500), 2)
    transaction_date = fake.date_time_between(start_date='-1y', end_date='now')

    transHistory_dict = {
        "'history_id" : history_id,
        "customer_id" : customer_id,
        "transaction_id" : transaction_id,
        "previous_balance": previous_balance,
        "new_balance" : new_balance,
        "transaction_date" : transaction_date
    }
    
    transHistory_list.append(transHistory_dict)

df = pd.DataFrame(transHistory_list)

df.head()
```
